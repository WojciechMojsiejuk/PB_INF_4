```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
import matplotlib
import itertools

from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import preprocessing
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
from sklearn.ensemble import RandomForestClassifier

matplotlib.use('Agg')

%matplotlib inline
```

```python
#Liczba epok
iteracje = 10
```

```python
#Odczytanie pliku
baza = pd.read_csv("master.csv")
```

```python
baza
```

```python
#Zamiana bazy danych na machine friendly liczby
x, y = pd.factorize(baza['country'])
baza['country'] = x
x, y = pd.factorize(baza['sex'])
baza['sex'] = x
x, y = pd.factorize(baza['age'])
baza['age'] = x
x, y = pd.factorize(baza['country-year'])
baza['country-year'] = x
x, y = pd.factorize(baza['HDI for year'])
baza['HDI for year'] = x

#Zamiana wyjścia na gorące jedynki
lb = LabelBinarizer()
labels_categorical = lb.fit_transform(baza['generation'])
```

```python
baza
```

```python
labels_categorical
```

```python
lb.classes_
```

```python
baza = baza.drop(['generation'], axis = 1)
```

```python
#Normalizacja
x = baza.values

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
baza = pd.DataFrame(x_scaled)
```

```python
#Podział na dane treningowe i testowe
(x_train, x_test, y_train, y_test) = train_test_split(baza, labels_categorical, test_size=0.25, random_state=32)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# Ustalenie struktury sieci
model = Sequential()

model.add(Dense(100, activation='relu', input_shape=(11,)))
model.add(Dropout(0.3))

model.add(Dense(100, activation='softsign'))
model.add(Dropout(0.3))

model.add(Dense(100, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(len(y_train[0]), activation='softmax'))

model.summary() # Wyświetlenie podsumowania skonstruowanego modelu
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Trenowanie sieci
history = model.fit(x_train, y_train, batch_size=16, validation_data=(x_test, y_test), epochs=iteracje)
```

```python
# Ewaluacja
predictions = model.predict(x_test, batch_size=16)
print(classification_report(y_test.argmax(axis=1), predictions.argmax(axis=1), target_names=lb.classes_))
```

```python
arr = np.arange(0, iteracje)
plt.style.use("ggplot")
plt.figure()
plt.plot(arr, history.history["acc"], label="train_acc")
plt.plot(arr, history.history["val_acc"], label="val_acc")
plt.title("Training accuracy")
plt.xlabel("epoch #")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```

```python
arr = np.arange(0, iteracje)
plt.style.use("ggplot")
plt.figure()
plt.plot(arr, history.history["loss"], label="train_loss")
plt.plot(arr, history.history["val_loss"], label="val_loss")
plt.title("Training loss")
plt.xlabel("epoch #")
plt.ylabel("loss")
plt.legend()
plt.show()
```

```python
predictions = model.predict_classes(x_test)
```

```python
#Transformacja z kodowania gorącojedynkowego na liczby całkowite
labels_ints = np.array([np.where(label==1)[0][0] for label in y_test])
```

```python
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
cm = confusion_matrix(labels_ints, predictions)
```

```python
plot_confusion_matrix(cm, lb.classes_)
```
